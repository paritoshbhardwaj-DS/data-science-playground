```python
import numpy as np
stats = np.array([[1,2,3],[4,5,6]])
# print(stats)

#min
print(np.min(stats))

#min of each row
print(np.min(stats, axis = 1))
print(np.min(stats, axis = 0)) # between the two rows which is mininum

#max
print(np.max(stats))
print(np.max(stats, axis = 1))

#sum
print(np.sum(stats))
print(np.sum(stats, axis = 1)) # sum of 1+2+3 = 6 and 4+5+6 = 15
print(np.sum(stats, axis = 0)) # sum of 1+4 = 5 2+5 = 7
```
