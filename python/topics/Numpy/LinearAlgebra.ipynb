```python
import numpy as np
import matplotlib.pyplot as plt

# Define vectors (columns of matrix)
v1 = np.array([2, 0])
v2 = np.array([1, 3])

# Matrix
A = np.column_stack((v1, v2))

# Determinant
det = np.linalg.det(A)

# Parallelogram points
origin = np.array([0, 0])
points = np.array([
    origin,
    v1,
    v1 + v2,
    v2,
    origin
])

# Plot
plt.figure(figsize=(6,6))
plt.plot(points[:,0], points[:,1], 'b-')
plt.fill(points[:,0], points[:,1], alpha=0.3)

# Plot vectors
plt.arrow(0,0,v1[0],v1[1], head_width=0.2, color='red', label='v1')
plt.arrow(0,0,v2[0],v2[1], head_width=0.2, color='green', label='v2')

plt.text(1, -0.5, f"Determinant (Area) = {det:.1f}")
plt.axhline(0); plt.axvline(0)
plt.grid()
plt.gca().set_aspect('equal')

plt.title("Determinant as Area of Parallelogram")
plt.legend()
plt.show()
```

```python
from mpl_toolkits.mplot3d import Axes3D

# Define 3 vectors
v1 = np.array([1, 0, 0])
v2 = np.array([1, 2, 0])
v3 = np.array([0, 1, 3])

A = np.column_stack((v1, v2, v3))

det = np.linalg.det(A)

fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')

# Origin
origin = np.array([0,0,0])

# Plot vectors
ax.quiver(*origin, *v1, color='r')
ax.quiver(*origin, *v2, color='g')
ax.quiver(*origin, *v3, color='b')

ax.set_title(f"Determinant (Volume) = {det:.2f}")

ax.set_xlim([0,3])
ax.set_ylim([0,3])
ax.set_zlim([0,3])

plt.show()
```
