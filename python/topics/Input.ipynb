Input in same line

```python
m , n = map(int, input().strip().split())
print(m,n)
# strip() is used to remove leading and trailing whitespace
```

Input in different line

```python
m = int(input())
n = int(input())
print(m)
print(n)
```

Input without map()

```python
m , n = int(input().strip().split())
# this will give error as input().strip().split() returns a list of strings and int() can convert only one string at a time, not a list
```
