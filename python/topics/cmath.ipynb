The cmath module is a built‑in Python module used for complex number mathematics.

✅ Use cmath when your calculations involve imaginary numbers

✅ Complex numbers are written as: a+bj (where j = √−1 in pyton)

```python
# square root
import cmath
print(cmath.sqrt(-4))
```

constants

```python
print(cmath.pi)
print(cmath.e)
```

Common functions

```python
#square root
print(cmath.sqrt(-2))
#e raised to x
print(cmath.exp(2))
#natural log
print(cmath.log(2))
#sin / cos
print(cmath.sin(1))
print(cmath.cos(1))
#angle of complex number
print(cmath.phase(1))
#magniture
z = 3 + 4j
print(abs(z))
print(cmath.phase(z)) # angle in radians
```
