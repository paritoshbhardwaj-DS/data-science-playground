1) Infinite loop

```python
#infinite loop
import itertools

nums = itertools.count(3)
print(nums)
print(next(nums))
print(next(nums))
print(next(nums))
print(next(nums))

# you can add as many as print(next(nums)) coz its infinite loop
```

2) Use a loop with a break (because it’s infinite)

```python
import itertools

nums = itertools.count(1)

for x in nums:
    print(x)
    if x ==5:
        break
```

3) Use islice() to take first N values safely

```python
import itertools

nums = itertools.count(2)
print(list(itertools.islice(nums,5)))
```

itertools.product()  is one of the most useful tools in itertools. Think of it as a clean, efficient way to generate all possible combination pairs/tuples from multiple iterables—i.e., the Cartesian product (like nested for loops).


What does it do?
Given iterables A, B, C, it produces all tuples: (a, b, c) for every a in A, every b in B, every c in C

```python
# from itertools import product
# import itertools

A = [1,2]
B = ['a', 'b']
print(list(product(A,B)))
```

Equivalent to

```python
result = []
for a in A:
    for b in B:
        result.append((a, b))
```

repeat= defines the lenght

```python
print(list(product([0,1], repeat=3)))
```

```python
# How to generate all possible combination?
from itertools import product


colors = ["red", "blue"]
sizes = ["S", "M", "L"]
styles = ["round", "vneck"]

print(list(itertools.product(colors, sizes, styles)))
```

```python
# How to create grid of coordinated pairs?
from itertools import product

rows = range(2)
columns = range(3)

coords = list(product(rows,columns))
print(coords)
```

```python
# How to check from all the pairs a target pair?
from itertools import product

arr = [1,2,3]
target = 6

print(list(itertools.product(arr,arr)))

for a, b in product(arr, arr):
    if a+b == target:
        print(a,b)
```
