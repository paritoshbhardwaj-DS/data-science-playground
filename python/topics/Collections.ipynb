Collections.Counter in Python is a ready‑made frequency counter: it counts how many times each item appears in an iterable (list/string/tuple), or you can build it from a dictionary of counts.


It behaves like a dictionary where:

keys = items you’re counting
values = counts (integers)

```python
from collections import Counter
# C is capital

c = Counter([1, 2, 2, 3, 3, 3])
print(c)
print(c[1])
print(c[5])
```

1) From a list

```python
from collections import Counter
Counter(["a", "b", "a", "c", "b", "a"])
```

2) From a string

```python
Counter("banana")
```

3) from a dictionary

```python
Counter({"a": 5, "b": 2})
```

Count most appearing

```python
c = Counter("mississippi")
print(c.most_common(2))
```

```python
print(c.most_common())
# If you don’t pass n, it returns all in descending order:
```
