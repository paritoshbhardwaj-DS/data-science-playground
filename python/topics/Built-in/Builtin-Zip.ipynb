zip() pair items from multiple iterable position wise and returns an iterable of tuples
So if you have:

iterable1: items at positions 0,1,2…

iterable2: items at positions 0,1,2…

zip() produces:

(iterable1[0], iterable2[0])

(iterable1[1], iterable2[1])

(iterable1[2], iterable2[2])

…and so on.

```python
name = ["Aman", "Akar", "Anti"]
scores = [90, 89, 94]
z = zip(name, scores)
print(list(z))
```

```python
name = ["Aman", "Akar", "Anti"]
scores = [90, 89, 94]
city = ['blr', 'mum', 'kol']
z = zip(name, scores, city)
print(list(z))
```

ZIP stops at the shortest iterable

```python
name = ["Aman", "Akar", "Anti"]
scores = [90, 89]
z = zip(name, scores)
print(list(z))
```

If you still want to continue even if the values are missing

```python
from itertools import zip_longest
name = ["Aman", "Akar", "Anti"]
scores = [90, 89]
z = zip(name, scores)
print(list(zip_longest(name, scores, fillvalue=None)))
```

zip return an iterable and not a list

```python
name = ["Aman", "Akar", "Anti"]
scores = [90, 89, 94]
z = zip(name, scores)
print(z)            # its iterator and not a list
print(list(z))      #consumes it
print(list(z))      #not empty (already consumed)
```

use:
1) Loop through 2 lists together

```python
names = ["Aman", "Riya"]
scores = [90, 85]

for n, s in zip(names, scores):
    print(n, "scored", s)
```

2) create dictionary from keys+values

```python
keys = ["name", "age", "city"]
vals = ["Paritosh", 30, "Bangalore"]

d = dict(zip(keys, vals))
print(d)
```

3) unzip using *

```python
pairs = [("Aman", 90), ("Riya", 85), ("Karan", 88)]
names, score = zip(*pairs)
print(names)
print(score)
```
