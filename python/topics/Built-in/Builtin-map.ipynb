- map() applies a given function to each element of an iterable (like list, tuple, etc.) and returns an iterator (map object).
- map(function, iterable)
- function → The operation you want to apply
- iterable → List, tuple, etc.
- Output → map object (iterator)

```python
# without map
n = [1,2,3,4]
r = []

for i in n:
    r.append(i*2)
print(r)
```

```python
# with map
n = [1,2,3,4]
r = map(lambda x: x*2, n)
print(list(r))
```

```python
n = [1,2,3,4]
r = map(int, n)
print(list(r))
```

```python
n = [1,2,3,4]
m = [5,6,7,8]
r = map(lambda x, y: x+y, n, m)
print(list(r))
```

```python
#data cleaning
names = ['  John ', ' Alice ', ' Bob ']
clean = map(str.strip, names)
print(list(clean))
```

```python
#string operations
words = ['apple', 'banana']
print(list(map(str.upper, words)))
```

```python
print(list(map(len, ['apple', 'banana']))
```
