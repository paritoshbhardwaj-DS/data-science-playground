reduce() is a function that applies another function cumulatively on elements of an iterable (like a list) and reduces it to a single value

- reduce(function, iterable[, initializer])
- function takes 2 argument
- iterable - list/tuple etc
- initializer - optional(starting value)

```python
from functools import reduce
n = [1,2,3,4,5]
result = reduce(lambda x, y: x+y, n)
print(result)
```

internally it works like this
step 1- 1+2 = 3
step 2 -3+3 = 6
step 3 - 6+4 = 10
step 4 - 10+5 = 15

1. sum of list

```python
reduce(lambda x,y: x+y, [1,2,3])
```

2. product

```python
reduce(lambda x,y: x*y, [3,2,3])
```

3. find max

```python
reduce(lambda x, y: x if x>y else y, [1,2,3])
```

4. string concatenation

```python
reduce(lambda x,y: x + " " + y, ["hello", "world"])
```

5. with initializer

```python
reduce(lambda x,y: x+y, [1,2,3],10)
```
