- Returns
True → if at least one element is truthy
- Returns
False → if all elements are falsy or iterable is empty

```python
print(any([False, True, False]))
print(any([0, True, None]))
print(any([]))
```

1) check if number is positive

```python
nums = [-1, 9, 2, 3]
print(any(n>0 for n in nums))
```

2) check if string is empty

```python
words = ['apple', '', 'banana']
print(any(w == '' for w in words))
```

3) using with condition

```python
arr = [2,3,7]
print(any(x%2 ==0 for x in arr))
```

- any() --> at least ONE is true
- all() --> EVERY element is true

```python
print(any([1>0,1==0,1<0]))
print(any([1<0,2<1,3<2]))
print(all(['a'<'b','b'<'c']))
print(all(['a'<'b','c'<'b']))
```
