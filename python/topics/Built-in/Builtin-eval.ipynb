eval() is a built‑in Python function that takes a string and executes it as a Python expression, then returns the result.

Syntax:
eval(expression, globals=None, locals=None)


It can evaluate expressions
Expressions are things that produce a value, like:


- math: "10/2"
- comparisons: "5 > 3"
- list/dict literals: "[1,2,3]", "{'a':1}"


❌ It cannot directly run statements
import, def, class, for, while, assignments

1) Simple math

```python
print(eval("2 + 3 * 4"))
```

2) using variable

```python
x = 10
print(eval("x * 10"))
```

3) evaluating function

```python
print(eval("len([1,2,3,4,5])"))
```

Why eval() is dangerous (very important)

If you run eval() on user input, the user can execute harmful code.

Example (DO NOT DO THIS):

```python
user_input = input()
eval(user_input)
```

A malicious input could be something like:

- deleting files
- reading system data
- running OS commands
So the golden rule is:
- Never use eval() on untrusted input.
