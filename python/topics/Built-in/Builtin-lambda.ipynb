- What is lambda?
- Lambda = “a small anonymous function”
- A lambda is a quick way to create a function without naming it

1) Sort list of tuples by index 1

```python
arr = [(101, 50), (102, 20), (103, 40)]
print(sorted(arr, key = lambda x : x[1]))
```

2) Sort in descending order

```python
arr = [(101, 50), (102, 20), (103, 40)]
print(sorted(arr, key = lambda x : x[1], reverse = True))
```

3) Sort by multiple keys (tie breaker)
- Sort by index 1, then by index 0:

```python
arr = [(2, 10), (1, 10), (3, 5)]
print(sorted(arr, key=lambda x: (x[1], x[0])))
```
