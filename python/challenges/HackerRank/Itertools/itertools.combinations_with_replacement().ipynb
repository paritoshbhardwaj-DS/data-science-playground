itertools.combinations_with_replacement(iterable, r) generates all possible combinations of length r where:order does NOT matter / elements CAN repeat

itertools.combinations_with_replacement(iterable, r)

iterable: input sequence (list, string, tuple, etc.)

r: size of each combination


Returns: an iterator of tuples

```python
from itertools import combinations_with_replacement
from itertools import combinations

print(list(combinations_with_replacement([1,2,3],2)))
print(list(combinations([1, 2, 3], 2)))
```

```python
from itertools import combinations_with_replacement
print(list(combinations_with_replacement('12345',2)))
```

```python
A = [1,1,3,3,3]
print(list(combinations(A,2)))
```

```python
from itertools import combinations_with_replacement
a = input().split()
s = sorted(a[0])
r = int(a[1]) #2
sorted_list = []

for p in combinations_with_replacement(s,r):
    sorted_list.append(''.join(p))
for items in sorted(sorted_list):
    print(items)
sorted_list.clear()
```
