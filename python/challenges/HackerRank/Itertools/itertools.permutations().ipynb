itertools.permutations(iterable[, r]) generates all possible ordered arrangements (permutations) of elements from an iterable. Order matters in permutations.

itertools.permutations(iterable, r=None)

- iterable: any iterable (list, string, tuple, range, etc.)
- r (optional): length of each permutation tuple. If r is not given, it defaults to the length of the iterable (n)

It returns an iterator producing tuples, each tuple being one permutation.

```python
from itertools import permutations
p = permutations([1,2,3],2)
print(list(p))
```

```python
from itertools import permutations
p = permutations(range(100), 2)  # huge, but still safe as iterator
print(next(p))                   # gives first tuple
```

```python
from itertools import permutations
a = input().split()
s = a[0]
r = int(a[1])
sorted_list = []
for p in permutations(s,r):
    sorted_list.append(''.join(p))

for items in sorted(sorted_list):
    print(items)

```
