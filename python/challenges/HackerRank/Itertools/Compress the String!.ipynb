itertools.groupby(iterable[, key]) is used to group consecutive items in an iterable that share the same “key”. It’s super handy for things like run-length encoding, chunking logs by status, or grouping already-sorted data.

groupby looks at the list from left to right and groups SAME numbers that are NEXT TO EACH OTHER.


It does NOT look at the whole list at once.


It only cares about consecutive (continuous) values.

from itertools import groupby

groupby(iterable, key=None

iterable: any iterable (list, string, generator, file lines, etc.)

key: a function that maps each item to a group key

If key is None, the item itself is used as the key.

```python
from itertools import groupby

data = [1, 1, 2, 2, 2, 1]
for k, g in groupby(data):
    print(k, list(g))


# output:
# 1 [1, 1] -> value 1 appears as [1,1] consecutively
# 2 [2, 2, 2] -> value 2 appears as [2,2,2] consecutively
# 1 [1]
```

```python
from itertools import groupby

data = [1, 1, 2, 2, 2, 1]
data.sort()

for k, g in groupby(data):
    print(k, list(g))
```
