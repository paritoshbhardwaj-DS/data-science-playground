```python
from itertools import product

N = 3
K = 1000
A = [5,4]
B = [7,8,9]
C = [5,7,8,9,10]
D = list(product(A,B,C))


sum = 0

for items in D:
    # print(items)
    squared = [(x**2) for x in items]
    # print(squared)

    total = 0

    for h in squared:
        total +=h
    S = total % K
    # print(S)

    if sum < S:
        sum = S
print(sum)
```

```python
from itertools import product
Q , W = map(int, input().split())
nums = []
for i in range(Q):
    k = list(map(int, input().split()))
    arr = k[1:]
    nums.append(arr)


D = (list(product(*nums)))

sum = 0

for items in D:
    # print(items)
    squared = [(x**2) for x in items]
    # print(squared)

    total = 0

    for h in squared:
        total +=h
    S = total % W
    # print(S)

    if sum < S:
        sum = S
print(sum)
```
