itertools.combinations(iterable, r) generates all possible selections of length r where order does NOT matter and no element is repeated within a selection.

Key idea: Choose r items from iterable; (a, b) is the same as (b, a).

itertools.combinations(iterable, r)

iterable: any iterable (list, tuple, string, range, etc.)


r: size of each combination (not optional)


Returns: an iterator of tuples


```python
from itertools import combinations
print(list(combinations([1,2,3],2)) # only one of (1,2) and (2,1) included coz order does not matter
```

```python
from itertools import combinations
a = input().split()
s = sorted(a[0]) # this is done so as to have the inputs in order
r = int(a[1]) #2
sorted_list = []
for i in range(1,r+1,1):
    for p in combinations(s,i):
        sorted_list.append(''.join(p))
    for items in sorted(sorted_list):
        print(items)
    sorted_list.clear()




```
