```python
# lets combine everything
def merge_the_tools(string, k):

    l = len(string)
    non_unique = []
    for i in range(0,l,k):
        non_unique.append((string[i:i+k]))

    for p in non_unique:
        unique = []
        unique1 = []
        for a in range(k):
            unique = list(p[a])
            for b in range(k):
                if unique == p[b]:
                    if p[b] not in unique1:
                        unique1.append(p[b])
                else:
                    if p[b] not in unique1:
                        unique1.append(p[b])
        result = ''.join(unique1)
        print(result)

if __name__ == '__main__':
    string, k = input(), int(input())
    merge_the_tools(string, k)
```
