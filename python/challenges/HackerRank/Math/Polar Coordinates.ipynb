You are given a complex . Your task is to convert it to polar coordinates.

Complex number z = x + yj

Polar form is z = r(cosθ + isinθ)

where r = magnitude(distance from origin)

θ(theta) = angle(direction), in radians


```python
import cmath
z = 3 + 4j
r, theta = cmath.polar(z)
print("Magnitude:", r)
print("Angle (radians):", theta)
```

What is the phase (argument)?

The phase (also called the argument) is:

✅ The angle the complex number makes with the positive real axis. It tells you which direction the number is pointing.

z = 3 + 4i
tan(θ) = 4/3
θ = 53.13 degree

This means the number points 53° above the real axis.

```python
import cmath
z = 3 + 4j
print(cmath.phase(z))
```

Convert to degrees:

```python
import math
print(math.degrees(cmath.phase(z)))
```

Why is it called “argument”?

Because it argues the direction of the complex number.

Magnitude → size

Argument → direction

What is modulus of a complex number?
The modulus (also called absolute value) is:

✅ The distance of the complex number from the origin (0, 0)

It is written as: ∣z∣
