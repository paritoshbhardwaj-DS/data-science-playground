```python
a =4
# order in asending order
total_inc = 0
for i in range(1,a+1):
    sum = i* (pow(10,(a-i)))
    total_inc +=sum

#order in descending order
total_des = 0
for j in range(1,a):
    sum1 = int(j* (pow(10,(j-1))))
    total_des +=sum1
print(str(total_inc)+str(total_des))

          # + str((j* (pow(10,(j-1))) for j in range(1,a))))
```
