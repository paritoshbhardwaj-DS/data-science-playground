Kevin and Stuart want to play the 'The Minion Game'.

Game Rules

Both players are given the same string, .
Both players have to make substrings using the letters of the string .
Stuart has to make words starting with consonants.
Kevin has to make words starting with vowels.
The game ends when both players have made all possible substrings.

Scoring
A player gets +1 point for each occurrence of the substring in the string .

but this returns the output as tuple and not string

why ? because combination is conceptually a set/list/tuple of items
even if the item has only one item it will still be a tuple


```python
#Lets try to convert that into a list so that we can iterate

from itertools import combinations

name = 'BANANA'

list_combo = []

print(list)
for combo in combinations(name, 1):
    list_combo.append(list(combo))
for combo in combinations(name, 2):
    list_combo.append(list(combo))
for combo in combinations(name, 3):
    list_combo.append(list(combo))
print(list_combo)
```

But this is giving us duplicate results

```python
from itertools import combinations

name = 'BANANA'

uni_combo = set()

for r in (1,2,3):
    for combo in combinations(name,r):
        uni_combo.add(tuple(combo))
print(uni_combo)
```

set() helps in creating unique items and removing duplicate items
Think of a set like a bag where: If the same item comes again, it will NOT be added twice

```python
items = ['A', 'A', 'B', 'B', 'B']
s = set(items)
print(s)
```
