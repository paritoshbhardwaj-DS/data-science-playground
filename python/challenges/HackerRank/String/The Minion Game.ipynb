The code you've written is logically sound for small words, but it will crash on the larger "hidden" test cases (like Test Case #4) because of how it handles memory and time.
Using string[i:j]: Slicing a string creates a new string in memory. If you do this inside a loop for \(10^{6}\) characters, your memory will explode.
Using .count(): If you find a substring and then call string.count(substring), you are scanning the entire massive string over and over. This turns your code into \(O(N^{2})\), which is death for Case #4.
Storing substrings in a list/set: sub_strings.append(s[i:j]) will consume gigabytes of RAM.
