```python
Given an integer,n, print the following values for each integer i from 1 to n:
1. Decimal
2. Octal
3. Hexadecimal (capitalized)
4. Binary
https://www.hackerrank.com/challenges/python-string-formatting/problem?isFullScreen=true
```

```python
# print decimal with left adjusted
for i in range(5):
    print(str((i+1)).rjust(len(format(5, 'b'))), ' ')
```

```python
# print octal 1.1
for i in range(2):
    print((format(i+1, 'o')).ljust(len(format(2, 'b'))),'*')

# l = len(str(format(2, 'b')))
# s = "p".ljust(l,'*')
# print(s)
```

```python
# print only integer part of octal
# format(number, 'o') converts the number to a string in octal.
# 'o' is the format specifier for octal.
# why jump form 7-10 . coz after 7 the next number is 10 in octal
for i in range(5):
    print(format(i+1, 'o'))
```

```python
# print only hexadecimal
for i in range(5):
    print((format(i+1, 'X')).rjust(len(format(5, 'b'))), ' ')
```

```python
# print only binary with right adjusted
for i in range(5):
    print((format(i+1, 'b')).rjust(len(format(5, 'b'))), ' ')
```

```python
# print automatically introduces space so ' ' not required
print('a','b','c','d')
print('a',' ','b',' ','c',' ','d')
```

```python
# final combined code
def print_formatted(number):
    for i in range(n):
        print((str(i+1).rjust(len(format(n, 'b')), ' '))
              ,(format(i+1, 'o').rjust(len(format(n, 'b')), ' ')
               ,(format(i+1, 'X').rjust(len(format(n, 'b')), ' '))
               ,(format(i+1, 'b').rjust(len(format(n, 'b')), ' ')))

if __name__ == '__main__':
    n = int(input())
    print_formatted(n)
```

```python
# refined
# we can define the width prior
def print_formatted(number):

# width based on binary length of number
    width = len(format(number, 'b'))

#use number instead of n as good practise
    for i in range(number):
# no need to define the filler ' ' if its only one space
        print(str(i+1).rjust(width)
              ,format(i+1, 'o').rjust(width)
               ,format(i+1, 'X').rjust(width)
               ,format(i+1, 'b').rjust(width))

if __name__ == '__main__':
    n = int(input())
    print_formatted(n)
```
