```
# split() and split(’ ‘ )
# split() S
# 1. splits on any whitespace (space, tab, newline)
# 2. removes any extra spaces
s = "paritosh    bhardwaj"
print(s.split())
# Extra spaces are lost forever
```

```
# split(' ')
# Splits only on a single space and preserves extra spaces
s = "paritosh    bhardwaj"
print(s.split(' '))
```

```
# capitalize() vs upper()
# capitalize() - Makes only the FIRST character uppercase
# upper() - Converts ALL letters to uppercase
"paritosh".capitalize()
```

```
# (no change, first char isn’t a letter)
"1abc".capitalize()
```

```
"paritosh".upper()
```

```
# Complete the solve function below.

def solve(s):
    name_list = list(s.split(' '))
    new = []
    for name in name_list:
        new_name = name.capitalize()
        new.append(new_name)
    new_list = ' '.join(new)
    return new_list
```

```
solve('1 w 2 r 3g')
```

```
solve('chris alan')
```

```
solve('pArItOsH bHaRdWaJ')
```
