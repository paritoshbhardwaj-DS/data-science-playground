Python counts weekdays like this:
0 = Monday
6 = Sunday

```python
1) Print a month calendar
```

```python
import calendar
print(calendar.month(2026,4))
```

```python
# Print full calendar
print(calendar.calendar(2026))
```

Weekday setting. By default, calendars start on Monday.
You can change globally:

```python
import calendar
calendar.setfirstweekday(calendar.SUNDAY)
print(calendar.month(2026, 4))
print(calendar.firstweekday())
```

What day of the week is this date?

```python
import calendar
wd = calendar.weekday(2026, 4, 22)
print(wd)
```

How many days in a month?

```python
import calendar
first_wd, num_days = calendar.monthrange(2026,4)
print(first_wd, num_days)
```

Leap year helpers

```python
import calendar
print(calendar.isleap(2024))
print(calendar.isleap(2023))
print(calendar.leapdays(2000, 2050))
# Count of leap years in [2000, 2050)
```

Get the month as matrix (weeks as lists)

Returns a list of weeks; each week is a list of 7 ints.
Days outside the month are 0.

```python
import calendar
weeks = calendar.monthcalendar(2026,4)
print(weeks)
for w in weeks:
    print(w)
```
