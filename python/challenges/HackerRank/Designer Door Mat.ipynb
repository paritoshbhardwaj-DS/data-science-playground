Mr. Vincent works in a door mat manufacturing company. One day, he designed a new door mat with the following specifications: Mat size must be X. ( is an odd natural number, and  is  times .) The design should have 'WELCOME' written in the center. The design pattern should only use |, . and - characters.

Tools required:
**.ljust(width, '_')** -> This method returns a left aligned string of length width. and rest as filler.
**.center(width)** -> This method returns a centered string of length width.
**.rjust(width)** -> This method returns a right aligned string of length width.

```python
width = 20
print('paritosh'.ljust(width,'*'))
```

```python
width = 20
print('paritosh'.rjust(width,'*'))
```

```python
width = 20
print('paritosh'.center(width,'*'))
```
