A set is an unordered collection of elements without duplicate entries.
When printed, iterated or converted into a sequence, its elements will appear in an arbitrary order. Basically, sets are used for membership testing and eliminating duplicate entries.

```python
print(set([1,2,3]))
print(set('paritosh bhardwaj'))
print(set(['paritosh', 'paritosh', 'bhardwaj', '123', '123']))
# no duplicate / no order
```

Task
Now, let's use our knowledge of sets and help Mickey.

Ms. Gabriel Williams is a botany professor at District College. One day, she asked her student Mickey to compute the average of all the plants with distinct heights in her greenhouse.
Formula -
Average = (sum  of distinct heights) / (total number of distinct heights)

```python
def average(array):
# sum of elements of arrays allowed
    result = float(sum(set(array)))/len(set(array))
    return(result)

if __name__ == '__main__':
    n = int(input())
    arr = list(map(int, input().split()))
    result = average(arr)
    print(result)
```
