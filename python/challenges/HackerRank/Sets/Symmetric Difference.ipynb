Sets are an unordered collection of unique values. A single set contains values of any immutable data type.

Creating Sets:

```python
#directly assigning value to set
myset = {1,2}
print(myset)

# initializing a set
myset = set()
myset = set(['a','b','c','d','e','f'])
print(myset)
```

Modifying set - add() function

```python
myset.add('c')
print(myset)
# if an element already exists then nothing happens
myset.add('a')
print(myset)
myset.add((5,4))
print(myset)
```

Update()

```python
#1 update() only works for iterable objects
myset.update([1, 2, 3, 4])
print(myset)
# list (iterable)
# Python loops through the list and add each element:
# add 1 2 3 4 to the original list
# List is ordered, but sets are unordered, so the printed order is arbitrary.
```

```python
#2
myset.update({10,11})
print(myset)
# Python loops through the list and add each element:
# add 1 already there , adds 10,11
```

```python
#3
myset.update({1, 6}, [5, 13])
# two iterable
# a set {1,6}
# a list [5,13]
# equivalent to

# for x in {1, 6}:
#     myset.add(x)

# for y in [5, 13]:
#     myset.add(y)
print(myset)
```

There is a difference between add() and update() - Add() adds only one element / update(iterable) add many elements from an iterable

```python
myset.update('xyz')
print(myset) # add x, y, z coz a string is iterable
myset.add('xyz')
print(myset) # adds xyz as one element
```

Removing items - Both discard() and remove() takes a single value as an argument and removes that value from the set. discard() is the value is not present then it does nothing. remove() will give error

```python
myset.discard(23)
print(myset)
```

```python
myset.remove(23)
print(myset)
```

Common set operations:
union() values which exist in a or b

```python
a = {1,2,3}
b = {3,4,5}
c = a.union(b)
print(c)
```

intersection()

```python
a = {1,2,3}
b = {3,4,5}
c = a.intersection(b)
print(c)
```

difference() Values which exist in a but not in b

```python
a = {1,2,3}
b = {3,4,5}
c = a.difference(b)
print(c)
```

```python
# union() and intersection() are symmetric methods
a.union(b) == b.union(a)
```

```python
a.intersection(b) == b.intersection(a)
```

```python
a.difference(b) == b.difference(a)
```

```python
Task
Given 2 sets of integers, M and N, print their symmetric difference in ascending order. The term symmetric difference indicates those values that exist in either M or N but do not exist in both.
```
