```python
cube = lambda x: x*x*x

def fibonacci(n):
    a = []
    if n >=2:
        f = [0,1]
        for i in range(n-2):
            a = f[i]+f[i+1]
            f.append(a)
        return(f)
    elif n == 0 :
        f = []
        return(f)
    elif n ==1:
        f = [0]
        return(f)
if __name__ == '__main__':
    n = int(input())
    print(list(map(cube, fibonacci(n))))

#in case of n = 0 , the expected output is null []
```
