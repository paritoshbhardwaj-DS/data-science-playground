```python
# N = students
# X = subject
M = list()
N, X = map(int, input().split())
for a in range(X):
    M.append(list(map(float, input().split())))
T = list(zip(*M))
# print(T)

for a in T:
    sum = 0
    for b in a:
        sum = sum + b
    print(round(float((sum)/X),1))
```
