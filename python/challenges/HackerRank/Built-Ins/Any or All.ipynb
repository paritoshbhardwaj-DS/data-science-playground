```python
N = int(input())
M = map(str, input().split())
print(any(a == ''.join(reversed(a)) for a in M))
```

```python
N = int(input())

M = list(map(int, input().split()))

c1 = all(x>0 for x in M)
c2 = any(str(a) == str(a)[::-1] for a in M)
print(c1)
print(c2)
print(c1 and c2)
```

errors: 1
- M = map(int, input().split()) is incorrect. M is a map iterator, and once you iterate over it (in all(...)), it gets consumed → nothing left for any(...).
-M = map(int, input().split())
- c1 = all(x > 0 for x in M)     # consumes M fully ✅
- c2 = any(... for a in M)       # now M is empty ❌ -> any(empty) = False


error2:
- str(a) == str(a[::-1]) is incorrect as a is not str
- any(str(a) == str(a)[::-1] is correct
