```python
# K = 'Sorting1234'
K = str(input())
lower = sorted(i for i in K if i.islower())
upper = sorted(i for i in K if i.isupper())
odd = sorted(i for i in K if i.isdigit() and int(i)%2 != 0)
even = sorted(i for i in K if i.isdigit() and int(i)%2 == 0)
result = "".join(lower+upper+odd+even)
print(result)
```
