```python
#alternate
def fun(s):
    try:
        username, rest = s.split("@")
        website, ext = s.split(".")
    except ValueError:
        return False

    if not username or not website or not ext:
        return False

    if not all(c.isalnum() or c in "-_" for c in username):
        return False

    if not website.isalnum():
        return False

    if not ext.isalpha() or len(ext) >3:
        return False

    return True

fun("paritosh.bhard@gmai.com")
```
