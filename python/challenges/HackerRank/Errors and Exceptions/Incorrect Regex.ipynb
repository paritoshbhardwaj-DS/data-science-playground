A valid regex is a pattern that follows the syntax rules of a regex engine (for example Python’s re module) and can be successfully compiled and executed without errors.

What makes a regex valid (in Python)
A regex is valid if:

✅ All brackets, parentheses, and braces are properly closed

✅ Special characters are used in correct contexts

✅ Quantifiers (* + ? {}) are applied to something valid

✅ Escapes (\d, \s, \w, etc.) are correct

✅ The pattern can be compiled by re.compile() without error


```python
import re

n = int(input())

for _ in range(n):

    s = input()
    try:
        re.compile(s)
        print('True')
    except re.error as e:
        print('False', e)
```

```python
import re

def is_valid_regex(s):
    try:
        re.compile(s)
        return True
    except re.error:
        return False


print(is_valid_regex(r'.*\+'))
print(is_valid_regex(r'.*+'))
```
