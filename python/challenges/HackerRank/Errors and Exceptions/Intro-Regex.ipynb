Regex (short for regular expression) is a pattern language used to search, match, extract, split, or replace text. In Python, regex is provided through the built-in re module.

Think of regex as a powerful “find” feature where you can describe rules like:

“Find any email address”
“Find all numbers with 10 digits”
“Find dates like DD-MM-YYYY”
“Replace multiple spaces with one space”
“Extract everything between brackets”

Understanding regex patterns (the basics)
Here are some of the most used tokens:
Character types

\d → digit (0–9)

\w → word char (letters, digits, underscore)

\s → whitespace (space, tab, newline)

. → any character (except newline by default)

Quantifiers (how many)

* → 0 or more
+ → 1 or more
? → 0 or 1
{m} → exactly m
{m,n} → between m and n

Anchors (position)

^ → start of string
$ → end of string
\b → word boundary

Character sets

[abc] → a or b or c
[a-z] → any lowercase letter
[^0-9] → anything except digits

 1) re.search() — find the first match anywhere

```python
import re
text = "Order ID: 98765 12344"
m = re.search(r"\d+", text)
print(m.group(0))
print(m.start())
print(m.span())
```

Step by step explanation:

import re --> # import regular expression module

m = re.search(r"\d+", text) --> search the entire string from left to right/ it returns                                          the first match it finds/ if it finds nothing

r".."  -> raw string (python doesnt treat \d as escape)
\d     -> means digit (0-9)
+      -> means one or more times
\d+    -> means match the continuous sequence of digits
"Order ID: 98765" - first continous digit sequence is 98765
So now m is a match object. Not a string - it contains what was matched, where it was matched (start/end),
group() --> returns the entire matched text

2) re.findall() — return all matches as a list

```python
import re
text = "Prices: 10, 20, 35"
nums = re.findall(r"\d+", text)
print(nums)
```

3) re.match() — match only at the start of the string

```python
import re
print(bool(re.match(r"\d+", "1abc")))
print(bool(re.match(r"\d+", "abc123")))
```

```python
m = re.match(r"\d+", "123abc")
if m is not None:
    print(True)
else:
    print(False)
```

4) re.sub() — replace pattern with something else

```python
import re
text = "Hello    world   !!"
clean = re.sub(r"\s+", " ", text)
print(clean)
```

5) re.split() — split using a pattern

```python
import re
text = "a,b; c|d"
parts = re.split(r"[,:;|]\s*", text)
print(parts)
```

Examples

Extract a 10-digit phone number

```python
import re
text = "Call me on 9876543210 tomorrow"
print(re.findall(r"\b\d{10}\b", text))
```

validate a simple pincode (re.fullmatch)

re.fullmatch() is useful when you want the entire string to match the pattern, not just a part.

```python
import re
pin = "560001"
print(bool(re.fullmatch(r"\d{6}", pin)))
```

re.findall() -> Extract words starting with “A” or “a”

```python
import re
text = "Analytics and AI are amazing"
print(re.findall(r"\b[aA]\w+", text))
```
