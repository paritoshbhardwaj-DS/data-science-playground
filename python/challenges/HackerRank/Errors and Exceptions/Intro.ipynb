Errors detected during execution are called exceptions.


1) ZeroDivisionError - This error is raised when the second argument of a division or modulo operation is zero.

```python
a = 1
b = 0
print(a/b)
```

2. ValueError - This error is raised when a built-in operation or function receives an argument that has the right type but an inappropriate value.

```python
a = '1'
b = '#'
print(int(a)/int(b))
```

3) Handling Exceptions - The statements try and except can be used to handle selected exceptions. A try statement may have more than one except clause to specify handlers for different exceptions.

```python
#Code
try:
    print(1/0)
except ZeroDivisionError as e:
    print("Error Code:",e)
```

1) try is used to monitor the code that might fail
1/0 causes a division by zero
Python raises an exception instead of crashing the program immediately

2) ZeroDivisionError is a built in exception class, it occurs when we divide by 0
Python jumps here only if the specific error happens

3) as e
e is a variable, it stores the actual error object
e --> integer division or modulo by zero

Python tries to execute 1 / 0
It fails → raises ZeroDivisionError
Control moves to except
Error message is captured in e
Program prints a friendly message instead of crashing
