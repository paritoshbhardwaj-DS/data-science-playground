Example 2:
Create a class Rectangle with:
- Attributes : length, width
- Method     : area()        → length * width
- Method     : perimeter()   → 2 * (length + width)
- Method     : is_square()   → True if length == width

r1 = Rectangle(5, 5)
print(r1.is_square())   # True

```python
class Rectangle:
    def __init__(self, length, width):
        self.length = length
        self.width = width
    def area(self):
        return self.length * self.width
    def perimeter(self):
        return 2* (self.length + self.width)
    def is_square(self):
        return self.length == self.width


# Test 1 — Square
r1 = Rectangle(5, 5)
print(r1.area())        # 25
print(r1.perimeter())   # 20
print(r1.is_square())   # True

# Test 2 — Rectangle
r2 = Rectangle(4, 7)
print(r2.area())        # 28
print(r2.perimeter())   # 22
print(r2.is_square())   # False
```

Create a class Counter with:
- Attribute  : count = 0
- Method     : increment()  → count + 1
- Method     : decrement()  → count - 1
- Method     : reset()      → count = 0
- Method     : value()      → return current count

c = Counter()
c.increment()
c.increment()
c.decrement()
print(c.value())   # 1

```python
class Counter:
    def __init__(self):
    # def __init__(self, count=0): # this also can be used
        # self.count = count
        self.count = 0
    def increment(self):
        self.count += 1
    def decrement(self):
        self.count -= 1
    def reset(self):
        self.count
    def value(self):
        return self.count

c = Counter()
c.increment()
c.increment()
c.decrement()
print(c.value())   # 1
```

Example 4:
Create a class BankAccount with:
- Attributes : owner, balance
- Method     : deposit(amount)    → add to balance
- Method     : withdraw(amount)   → deduct (if sufficient funds)
- Method     : statement()        → print owner and balance

acc = BankAccount("Paritosh", 1000)

acc.deposit(500)

acc.withdraw(200)

acc.statement()   # Paritosh | Balance: 1300

```python
class BankAccount:
    def __init__(self, owner, balance):
        self.owner = owner
        self.balance = balance
    def deposit(self, deposit):
        # self.deposit = deposit
        self.balance = self.balance + deposit
    def withdraw(self, withdraw):
        # self.withdraw = withdraw
        if withdraw > self.balance:
            print("insufficient balance")
        else:
            self.balance = self.balance - withdraw
    def statement(self):
        print(f"{self.owner} | Balance: {self.balance}")

acc = BankAccount("Paritosh", 1000)
acc.deposit(500)
acc.withdraw(200)
acc.statement()   # Paritosh | Balance: 1300
```

Exercise 5 — Dunder Methods
Create a class Box with:
- Attributes : length, width, height
- __init__   → store dimensions
- __str__    → "Box(l x w x h)"
- __add__    → add volumes of two boxes
- __eq__     → True if volumes are equal
- volume()   → length * width * height

b1 = Box(2, 3, 4)

b2 = Box(1, 4, 6)

print(b1)           # Box(2 x 3 x 4)

print(b1 + b2)      # 48  (24 + 24)

print(b1 == b2)     # True

```python
class Box:
    def __init__(self, length, width, height):
        self.length = length
        self.width = width
        self.height = height
    def volume(self):
        return self.length * self.width * self.height
    def __str__(self):
        return f"Box({self.length} x {self.width} x {self.height})"
    def __add__(self, other):
        return self.volume() + other.volume()
    def __eq__(self, other):
        return self.volume() == other.volume()


b1 = Box(2, 3, 4)
b2 = Box(1, 4, 6)
b3 = Box(5, 5, 5)

print(b1)           # Box(2 x 3 x 4)
print(b2)           # Box(1 x 4 x 6)

print(b1.volume())  # 24
print(b2.volume())  # 24

print(b1 + b2)      # 48  ← 24 + 24
print(b1 == b2)     # True  ← both volume 24
```

Exercise 6 — Temperature Converter
Create a class Temperature with:
- Attribute  : celsius
- Method     : to_fahrenheit()  → (celsius * 9/5) + 32
- Method     : to_kelvin()      → celsius + 273.15
- __str__    → "X °C"

t = Temperature(100)

print(t.to_fahrenheit())   # 212.0

print(t.to_kelvin())       # 373.15

print(t)                   # 100 °C

```python
class Temperature:
    def __init__(self, celsius):
        self.celsius = celsius
    def __str__(self):
        return f"{self.celsius} °C"
    def to_fahreheit(self):
        return ((self.celsius * (9/5)) + 32)
    def to_kelvin(self):
        return self.celsius + 273.15

t = Temperature(100)
print(t.to_fahreheit())
print(t.to_kelvin())
print(t)
```

Exercise 7 — Shape Inheritance
Create a parent class Shape with:
- Attribute  : color
- Method     : describe() → "I am a {color} shape"

Create child classes:
- Circle(Shape)    → has radius    → area(), circumference()
- Rectangle(Shape) → has l, w      → area(), perimeter()

c = Circle("red", 7)

r = Rectangle("blue", 4, 5)

print(c.describe())        # I am a red shape

print(c.area())            # 153.93

print(r.perimeter())       # 18

```python
# # Basic syntax:
class Parent:                           #Parent class
    def greet(self):
        print("Hello from Parent")
    def Child(Parent):                  #Child inherits from Parent
        pass
c = Child()
c.greet()
```

```python
import math

# Parent class
class Shape:
    def __init__(self, color):
        self.color = color
    def describe(self):
        return f"I am a {self.color} shape"
#Child class 1
class Circle(Shape):
    def __init__(self, color, radius):
        super().__init__(color)             #calls parent __init__
        self.radius = radius

    def area(self):
        return round(math.pi * self.radius ** 2, 2)
    def circumference(self):
        return round(math.pi * 2 * self.radius, 2)
#Child class 2
class Rectangle(Shape):
    def __init__(self, color, length, width):
        super().__init__(color)
        self.length = length
        self.width = width
    def area(self):
        return self.length * self.width
    def perimeter(self):
        return 2 * (self.length * self.width)


# Testing
c = Circle("red", 7)
r = Rectangle("blue", 4, 5)

print(c.describe())         # I am a red shape     ← inherited from Shape ✅
print(c.area())             # 153.94
print(c.circumference())    # 43.98

print(r.describe())         # I am a blue shape    ← inherited from Shape ✅
print(r.area())             # 20
print(r.perimeter())        # 18



c = Shape("red")
print(c.describe())
```

Exercise 8 — Employee & Manager

Create class Employee with:

- Attributes : name, salary

- Method     : details() → print name and salary

Create class Manager(Employee) with:

- Extra attr : department

- Override   : details() → print name, salary, department

e = Employee("Raj", 50000)

m = Manager("Priya", 80000, "Analytics")


e.details()   # Raj | Salary: 50000

m.details()   # Priya | Salary: 80000 | Dept: Analytics

```python
class Employee:
    def __init__(self, name, salary):
        self.name = name
        self.salary = salary
    def details(self):
        print(f"{self.name} | Salary: {self.salary}")

class Manager(Employee):
    def __init__(self, name, salary, department):
        super().__init__(name, salary)
        self.department = department
    def details(self):
        print(f"{self.name} | {self.salary} {self.department}")

# Testing
e = Employee("Raj",   50000)
m = Manager("Priya", 80000, "Analytics")

e.details()   # Raj   | Salary: 50000
m.details()   # Priya | Salary: 80000 | Dept: Analytics
```
