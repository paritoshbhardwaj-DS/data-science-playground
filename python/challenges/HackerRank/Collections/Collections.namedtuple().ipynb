collections.namedtuple() is a factory function in Python that creates a lightweight class (a subclass of tuple) whose elements can be accessed both by index and by attribute name.

```python
from collections import namedtuple

Point = namedtuple("Point", ["x", "y"])
p = Point(10, 20)

print(p.x)  # 10
print(p.y)  # 20
print(p[0]) # 10 (still works like a tuple)
```

Readable (use .x instead of [0])
Immutable (like tuples; values can’t be changed)
Memory efficient (lighter than a regular class)
Tuple features (iteration, unpacking, indexing)

```python
from collections import namedtuple
Point = namedtuple('Point', 'x,y')
pt1 = Point(1,2)
pt2 = Point(3,4)
dot_product = (pt1.x * pt2.x) + (pt1.y * pt2.y)

print(dot_product)
```

```python
from collections import namedtuple
students = namedtuple("students", ["ID", "marks", "classes" ,"name"])

N = int(input())
students_list = []
for _ in range(N):
    row = input().split()
    e = students(*row)
    students_list.append(e)
marks = 0
for s in students_list:
    marks = marks + int(s.marks)
print(marks/N)
# print(Counter(students))
```

```python
from collections import namedtuple

N = int(input().strip())
headers = input().split()

students = namedtuple("students", headers)

total = 0
for _ in range(N):
    s = students(*input().split())
    total += int(s.MARKS)

print(total / N)
```
