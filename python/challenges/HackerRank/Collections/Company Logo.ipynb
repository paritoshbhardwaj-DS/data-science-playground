```python
from collections import Counter
from collections import deque

if __name__ == '__main__':
    s = input()

z = deque()
list = []
for items in s:
    list.append(items)
c = Counter(list)
d = c.most_common()
for i in range(3):
    z.append(d[i])
# print(z)
for key, values in z:
    print(key,values)
```

```python
from collections import Counter

if __name__ == '__main__':
    s = input()

c = Counter(s)
d = c.most_common()

def sort_key(items):
    return(-items[1], items[0])

sorted_list = sorted(d, key=sort_key)

for key,values in sorted_list[:3]:
    print(key,values)
```

sorted(d, key=sort_key) means sort the items in d, but don't compare the items directly.
First convert them into a sorting key using sort_key(item), then sort by that key

```python
# Ex1 - sorting numbers by distance from 10
nums = [12,3,9,21,10]
def sort_key(x):
    return abs(x-10)
sorted_list = sorted(nums, key=sort_key)
print(sorted_list)
```

```python
# Ex2: sorting words by length
words = ["apple", "kiwi", "banana", "fig"]
def sort_key(x):
    return len(x)
sorted_list = sorted(words, key=sort_key)
print(sorted_list)
```

```python
# Ex3: sorting only by values
d = [('b', 3), ('a', 2), ('c', 2), ('d', 1), ('e', 1)]
def sort_key(item):
    return(-item[1])
sorted_list = sorted(d, key=sort_key)
print(sorted_list)
# but in case of ties this will fail
```

```python
d = [('b', 3), ('a', 2), ('c', 2), ('d', 1), ('e', 1)]
def sort_key(item):
    return(-item[1], item[0])

# return (-item[1], item[0]) means it first gives preference to item[1] and then to item[0] in case of tie in item[1]

sorted_list = sorted(d, key=sort_key)
print(sorted_list)
```
