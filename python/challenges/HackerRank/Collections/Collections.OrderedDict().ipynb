OrderedDict is a dictionary-like container from the collections module that remembers the order in which keys were inserted

In older Python versions (< 3.7):

Normal dict did NOT guarantee insertion order.
So OrderedDict was used when order mattered.
In Python 3.7+ (and CPython 3.6):

Normal dict preserves insertion order by default, so in many cases you don’t need OrderedDict.

✅ But OrderedDict is still useful because it provides extra ordering-related features that normal dict doesn’t.

```python
from collections import OrderedDict

od = OrderedDict()
od["b"] = 2
od["a"] = 1
od["c"] = 3

print(od)
```

```python
for k in od:
    print(k)
```

```python
# Now (Python 3.7+), dict keeps order anyway:
d = {"b": 2, "a": 1, "c": 3}
print(d)
```

```python
# Feature A: move_to_end() - Move a key to the end(or beginning)

from collections import OrderedDict
od = OrderedDict([("a", 1), ("b", 2), ("c", 3)])
od.move_to_end("b")
print(od)

od.move_to_end("b", last=False)
print(od)
```

```python
# Feature B: popitem(last=True/False) - pop last/first item

from collections import OrderedDict

od.popitem()
print(od)
od.popitem(last=False)
print(od)
```

```python
# Feature C: order sensitivity

from collections import OrderedDict

od1 = OrderedDict([("a", 1), ("b", 2)])
od2 = OrderedDict([("b", 2), ("a", 1)])
print(od1 == od2)

# Normal dictionary ignores order in equality:
d1 = {"a": 1, "b": 2}
d2 = {"b": 2, "a": 1}
print(d1 == d2)  # True
```

```python
# HACKERRANK

# Option 1 (copilot)

from collections import OrderedDict
N = int(input())
d = OrderedDict()

for i in range(N):
    key, value = input().rsplit(" ", 1)
    value = int(value)
    d[key] = d.get(key, 0) + value
print(d)
```

```python
# Option2: Google

from collections import OrderedDict

net_prices = OrderedDict()
n = int(input())
for _ in range(n):
    item_data = input().split()
    price = int(item_data[-1])
    item_name = " ".join(item_data[:-1])

# if x in some_dict: (considers x to be key and not value)

    if item_name in net_prices:
        net_prices[item_name] += price
    else:
        net_prices[item_name] = price

for item_name, net_price in net_prices.items():
    print(f"{item_name} {net_price}")
```
