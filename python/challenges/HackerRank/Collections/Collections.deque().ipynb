collections.deque() is a double‑ended queue in Python. It’s designed for fast appends and pops from both ends, unlike lists which are slow at the front.

Deques support thread safe, memory efficient appends and pops from either side of the deque with approximately the same  performance in either direction.

```python
from collections import deque
d = deque([1,2,3])

# add element
d.append(4)       # add only one element to right
d.appendleft(5)   # add to left
d.pop()           # removes from right
d.popleft()       # removes from left
print(d[0])       # left most
print(d[-1])      # right most
d.clear()         # removes all elements
d.extend([6,7])   # adds multiple elements on right
d.extendleft([8,9])  # adds multiple elements on left
d.rotate(1)         # shifting all elements of a sequence to the right, where the last element                      comes to the front.
d.rotate(-1)      #rotates left
d.remove(7)       # removes first occurrence of value (from left to right)
d.rotate(3)        # rotate the deque to the RIGHT by 3 positions. [1, 2, 3, 4, 5] - [3, 4, 5,                                                                                          1, 2]
print(d)


# remove(value) → you specify the value
# pop() / pop(i) → you specify the position (list) or end (deque)
```

```python
from collections import deque

d = deque()

N = int(input())

for i in range(N):
    command = input().split()

    if command[0] == 'append':
        value = command[1]
        d.append(value)

    if command[0] == 'appendleft':
        value = command[1]
        d.appendleft(value)

    if command[0] == 'pop':
        d.pop()

    if command[0] == "popleft":
        d.popleft()

for items in d:
    print(items, end = " ")
```
