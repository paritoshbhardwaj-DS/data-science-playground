```python
from collections import defaultdict

n , m = map(int, input().split())

pos = defaultdict(list)

for i in range(1, n+1):
    word = input().strip()
    pos[word].append(i)


for _ in range(m):
    items = input().strip()
    if len(pos[items]) > 0:
        print(*pos[items])
    else:
        print(-1)
```
