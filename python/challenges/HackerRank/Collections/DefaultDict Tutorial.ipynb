```python
from collections import defaultdict

n , m = map(int, input().split())

pos = defaultdict(list)

for i in range(1, n+1):
    word = input().strip()
    pos[word].append(i)


for _ in range(m):
    items = input().strip()
    if len(pos[items]) > 0:
        print(*pos[items])
    else:
        print(-1)
```

But a defaultdict automatically creates a default value for missing keys.

```python
# Ex 1: Counting
from collections import defaultdict
count = defaultdict(int)

word = ('a','b','c','d')
for w in word:
    count[w] +=1
print(count)
```
