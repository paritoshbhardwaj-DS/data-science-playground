```python
from collections import deque

T = int(input())
for _ in range(T):
    n = int(input())
    d = deque(map(int, input().split()))

    top = float('inf')

    while d:
        left = d[0]
        right = d[-1]

        if left >= right and left <= top:
            top = d.popleft()
        elif right > left and right <= top:
            top = d.pop()
        else:
            print('No')
            break
    else:
        print('Yes')
```
