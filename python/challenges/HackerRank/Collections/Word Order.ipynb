```python
from itertools import permutations
for p in permutations("abcdefg", 1):
    print(''.join(p))
```

```python
from collections import OrderedDict

c = 0
od = OrderedDict()
N = int(input())

for _ in range(N):
    word = input().strip()

    if word in od:
        od[word] = od[word] + 1
    else:
        od[word] = c+1

print(len(od))
for word, value in od.items():
    print(f"{value}", end = " ")

```
